```python
import pandas as pd
movies = pd.read_csv('/content/drive/MyDrive/MovieAssignmentData.csv')
```

```python
movies.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. profit
movies.budget = round(movies['budget']/1000000,2)
movies.gross = round(movies['gross']/1000000,2)


```

```python
movies['profit'] = movies.gross - movies.budget
movies_by_profit = movies.sort_values("profit",ascending = False)
print(movies_by_profit)
```

```python
# 2. which language takes high profit

temp = movies.groupby('language')['profit'].mean().sort_values(ascending=False)
print(temp)

```

```python
# 3.which genres makes the most money?
temp = movies.groupby('genres')['profit'].mean().sort_values(ascending=False)
print(temp)
 
```

```python
# 4.Do the movies that have high vote_avergae make lot of profits ?
tempvar = movies.groupby('num_voted_users')['profit'].mean().sort_values(ascending=False)
print(tempvar)
```

```python
temp = pd.DataFrame({'best directors': movies['director_name']})
count_percentage = temp.value_counts()
print(count_percentage)

```

```python
# which producers generally do well?
tem = movies.groupby('director_name')['num_voted_users'].mean().sort_values(ascending=False)
print(tem)
```

```python
# finding the most popular genre
genre = movies.genres.str.split('|',expand=True).iloc[:,0:2]
genre.columns=['genre_1','genre_2']
genre.genre_2.fillna(genre.genre_1,inplace=True)
genre
```

```python
pd.concat([movies,genre], axis=1)
movies
```
